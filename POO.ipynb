```python
# Importer les paquages necessaire
import numpy as np
```

```python
# Créer une classe voiture

class Voiture:
    def __init__(self, marque, modele, annee, couleur):
        self.marque = marque
        self.modele = modele
        self.annee = annee
        self.couleur = couleur
        self.vitesse = 0

    def accelerer(self, vitesse_acceleration):
        self.vitesse += vitesse_acceleration

    def ralentir(self, vitesse_ralentissement):
        self.vitesse -= vitesse_ralentissement


```

```python
# Création des 5 objets

voiture1 = Voiture("BMW", "Série 3", 2023, "Rouge")
voiture2 = Voiture("Toyota", "Camry", 2022, "Bleu")
voiture3 = Voiture("Renault", "Clio", 2021, "Blanc")
voiture4 = Voiture("Peugeot", "208", 2023, "Noir")
voiture5 = Voiture("Audi", "A4", 2022, "Gris")
```

```python
# Affichage
voitures = [voiture1, voiture2, voiture3, voiture4, voiture5]
for v in voitures:
    print(f"Marque: {v.marque}, Modèle: {v.modele}, Année: {v.annee}, Couleur: {v.couleur}, Vitesse: {v.vitesse}")
```

# Encapsulation

```python
# Exemple d'encapsulation
class Personne:
    def __init__(self, nom, age):
        self._nom = nom
        self._age = age

    def get_nom(self):
        return self._nom

    def set_nom(self, nom):
        self._nom = nom

    def get_age(self):
        return self._age

    def set_age(self, age):
        if age >= 0:
            self._age = age
        else:
            print("L'âge ne peut pas être négatif.")

    def afficher_informations(self):
        print(f"Nom: {self._nom}, Age: {self._age}")


personne1 = Personne("Alice", 25)
personne1.afficher_informations()

print("Modification des attributs de la personne")

personne1.set_nom("Bob")
personne1.set_age(30)
personne1.afficher_informations()

#
```

# Héritage

```python
# Exemple d'héritage
class Vehicule:
    def __init__(self, marque, modele, annee):
        self.marque = marque
        self.modele = modele
        self.annee = annee
        self.vitesse = 0

    def accelerer(self, vitesse_acceleration):
        self.vitesse += vitesse_acceleration

    def ralentir(self, vitesse_ralentissement):
        self.vitesse -= vitesse_ralentissement

class Voiture(Vehicule):
    def __init__(self, marque, modele, annee, couleur):
        super().__init__(marque, modele, annee)
        self.couleur = couleur

# un objet
voiture1 = Voiture("Toyota", "Camry", 2022, "Bleu")
voiture1.accelerer(30)
print(f"Vitesse de la voiture: {voiture1.vitesse}")

#
```

# Polymorphisme

```python
# Exemple polymorphisme
class Animal:
    def __init__(self, nom):
        self.nom = nom

    def faire_son(self):
        pass

class Chat(Animal):
    def faire_son(self):
        return "Miaou"

class Chien(Animal):
    def faire_son(self):
        return "Wouf"

# Objets
chat = Chat("Chaton")
chien = Chien("Fido")

# Appel de la méthode faire_son()
print(chat.faire_son())
print(chien.faire_son())
#
```
