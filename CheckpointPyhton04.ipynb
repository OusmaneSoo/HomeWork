```python
# Définir la classe Account et ses attributs comme spécifié ci-dessus.
# Définir la méthode deposit(). Elle doit prendre un argument, le montant à déposer, et l'ajouter au solde du compte.
# Définissez la méthode withdraw(). Elle doit prendre un argument, le montant à retirer, et le soustraire du solde du compte. La méthode ne doit exécuter le retrait que si le solde du compte est supérieur ou égal au montant à retirer.
# Définir la méthode check_balance(). Elle doit renvoyer le solde actuel du compte.
# Créer une instance de la classe Compte et l'affecter à une variable appelée "mon_compte".
# Utiliser les méthodes de la classe pour déposer et retirer de l'argent du compte et vérifier le solde du compte.
# Tester le programme en créant plusieurs instances de la classe et en effectuant différentes transactions sur celles-ci.


class Account():
    def __init__(self, numéro_compte, balance, titulaire):
        self.numéro_compte  = numéro_compte
        self.balance = balance
        self.titulaire = titulaire

    def deposit(self, montant):
        self.balance += montant

    def withdraw(self, montant):
        if self.balance >= montant:
            self.balance -= montant
        else:
            print("Solde insuffisant")

    def check_balance(self):
        return self.balance

```

```python
mon_compte = Account("123456789", 35000, "Ousmane Sow")
mon_compte.deposit(12000)
mon_compte.withdraw(25000)
print(mon_compte.check_balance())
```
