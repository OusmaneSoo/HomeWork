```python
# Entrée:

# Entrez le premier nombre : 10

# Entrez le deuxième nombre : 5

# Entrez l'opérateur (+, -, *, /) :

n = input("Entrez le premier nombre : ")
m = input("Entrez le deuxième nombre : ")
op = input("Entrez l'opérateur (+, -, *, /) : ")
n = float(n)
m = float(m)

if op == "+":
  print(n + m)
elif op == "-":
  print(n - m)
elif op == "*":
  print(n * m)
elif op == "/" and m>0:
  print(n / m)

else:
  print("Erreur : opérateur non valide")
```

```python
# Ouvrez un nouveau fichier Python dans votre éditeur de code préféré.
# Définissez une fonction appelée "calculatrice" qui prend deux arguments, "num1" et "num2".
# Au sein de la fonction, demandez à l'utilisateur d'entrer un opérateur (+, -, *, /).
# Utilisez des instructions conditionnelles pour déterminer l'opération à effectuer en fonction de la saisie de l'utilisateur.
# Retournez le résultat de l'opération.
# Appeler la fonction calculatrice avec deux nombres et imprimer le résultat.
# Gérer les cas où l'utilisateur entre des opérateurs non valides ou essaie de diviser par zéro.

def Calculatrice(num1, num2):
  op = input("Entrez l'opérateur (+, -, *, /) : ")
  num1 = float(num1)
  num2 = float(num2)
  if op == "+":
    return num1 + num2
  elif op == "-":
    return num1 - num2
  elif op == "*":
    return num1 * num2
  elif op == "/" and num2 > 0:
    return num1 / num2
  else:
    return "Erreur : opérateur non valide"
quit = 'c'.lower()

while quit == "c".lower() or quit == "c".upper():
  num1 = input("Entrez le premier nombre : ")
  num2 = input("Entrez le deuxième nombre : ")
  print(Calculatrice(num1, num2))

  quit = input('tapez "C" pour continuer merci !')


```
